```python
from scapy.all import *
from scapy.layers.http import HTTP
import pandas as pd
import numpy as np
import csv
import os

load_layer("tls")
```

```python
BACKGROUND = "../background-trace/traces"
WEB1 = "web1-background.pcap"
WEB2 = "web2-background.pcap"
WEB3 = "web3-background.pcap"
WEB_TRACES = [WEB1, WEB2, WEB3]
OUTLOOK = "outlook-background.pcap"
TEAMS = "teams-background.pcap"
SKYPE = "skype-background.pcap"
OUT_FOLDER = "traces_out"
TRACES = "traces"

# Create output folder if it does not exist
if not os.path.exists(OUT_FOLDER):
    os.makedirs(OUT_FOLDER)
```

```python
for trace in WEB_TRACES:
    trace_path = os.path.join(BACKGROUND, trace)
    packets = rdpcap(trace_path)
    with open(os.path.join(OUT_FOLDER, f"{trace}.csv"), "w") as out_file:
        csv_writer = csv.writer(out_file)
        csv_writer.writerow(["ipsrc", "ipdst", "proto", "srcport", "dstport", "http_data", "ok", "malware"])
        for packet in packets:
            # empty strings
            ipsrc = ""
            ipdst = ""
            proto = ""
            srcport = ""
            dstport = ""
            http_data = ""
            if packet.haslayer("IP"):
                pass
                ipsrc = packet["IP"].src
                ipdst = packet["IP"].dst
                proto = packet["IP"].proto
            if packet.haslayer("TCP"):
                srcport = packet["TCP"].sport
                dstport = packet["TCP"].dport
            if packet.haslayer("UDP"):
                srcport = packet["UDP"].sport
                dstport = packet["UDP"].dport
            if packet.haslayer(HTTP):
                if packet[HTTP].haslayer("HTTP Request"):
                    http_data = packet[HTTP].Host.decode("utf-8")
            if packet.haslayer("TLS"):
                if packet['TLS'].type == 22:
                    if type(packet['TLS'].msg[0]) == scapy.layers.tls.handshake.TLSClientHello:
                        if packet['TLS'].msg[0].msgtype == 1:
                            http_data = packet['TLS']['TLS_Ext_ServerName'].servernames[0].servername.decode("utf-8")
                            # print(http_data)
            csv_writer.writerow([ipsrc, ipdst, proto, srcport, dstport, http_data, 1, 0])
```

```python
# mix the dataframes randomly, but keeping the order
# use a poisson distribution with lambda equal to the ratio of malware to ok
# to decide in which position to insert a bg packet into the malware df
ratio = len(malware_df) / len(web_df)
print(f"Ratio: {ratio}")
cur = 0
for i in range(len(web_df)):
    if i%1000 == 0:
        print(f"{i}/{len(web_df)}")
    cur += np.random.poisson(ratio)
    malware_df = pd.concat([malware_df.iloc[:cur], web_df.iloc[[i]], malware_df.iloc[cur:]]).reset_index(drop=True)
    cur += 1
malware_df
```

```python
# split labels and features
labels = malware_df[["ok" ,"malware"]]
features = malware_df.drop(columns=["ok", "malware"])
```

```python
# expanding ip addr object fields, separating bytes and converting them to float
for field_name, field_type in features.dtypes.items():
    if field_type == object:
        field_type = None
        for element in features[field_name]:
            # check if field is not a string
            if type(element) != str:
                continue
            # check if field is an ip address
            elif re.match(r"^(?:(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.){3}(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)$", element):
                field_type = 'ip'
                break

        if not field_type:
            print('Field type not found')
            print(field_name)
            break

        if field_type == 'ip':
            added_field = features[field_name].str.split(".", expand=True).astype('float64')

        features.drop(field_name, axis=1, inplace=True)
        features = pd.concat([features, added_field], axis=1)
features
```

```python
features.fillna("", inplace=True)
```

```python
# expand HTTP data
maxlen = 32

added_field = features["http_data"].str[:maxlen].str.encode('utf-8')

for i in range(len(added_field)):
    seq = [0] * maxlen
    arr = added_field[i]
    arrlen = len(arr)
    for j in range(arrlen):
        seq[maxlen - arrlen + j] = arr[j]
    added_field[i] = seq

for field in added_field:
    if field != list([0] * maxlen):
        print(field)

features.drop("http_data", axis=1, inplace=True)
features = pd.concat([features, pd.DataFrame(added_field.to_list())], axis=1)
features
```

```python
# normalize data
for column in features.columns:
    features[column] = features[column]  / features[column].abs().max()
features
```

```python
# add labels back
final_df = pd.concat([features, labels], axis=1)
final_df
```

```python
# write to file
final_df.to_csv(os.path.join(OUT_FOLDER, "malware_web_mixed_cleaned.csv"), index=False, header=False)
```
