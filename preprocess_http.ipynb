```python
TRACES = "traces"
OUT_FILE_ALL = "all.json"
OUT_FILE_NO_EMPTY = "no_empty.json"
get_params = list()
get_params_also_empty = list()
```

```python
# for each pcap file in the traces directory
# parse the trace and check if it is a HTTP GET request
# if it is, extract the URL parameters and add them to a list
for trace in os.listdir(TRACES):
    packets = rdpcap(os.path.join(TRACES, trace))
    for packet in packets:
        if packet.haslayer(HTTP):
            if packet[HTTP].haslayer("HTTP Request"):
                if packet[HTTP].Method == b"GET":
                    path = packet[HTTP].Path.decode("utf-8")
                    # check if the path has a query string
                    if "?" in path:
                        get_params.append(path.split("?")[1])
                        get_params_also_empty.append(path.split("?")[1])
                    else:
                        get_params_also_empty.append("")
# create labeled objects
get_params_labeled = list()
for param in get_params:
    get_params_labeled.append({"str": param, "attack": 1})
get_params_also_empty_labeled = list()
for param in get_params_also_empty:
    get_params_also_empty_labeled.append({"str": param, "attack": 1})
```

```python
# write labeled data to file
with open(OUT_FILE_ALL, "w") as f:
    json.dump(get_params_also_empty_labeled, f)
with open(OUT_FILE_NO_EMPTY, "w") as f:
    json.dump(get_params_labeled, f)
```

```python
print(f"number of GET requests: {len(get_params_also_empty_labeled)}\nnumber of GET requests with parameters: {len(get_params_labeled)}")
```
